```python
fig1 = px.violin(df, y=df["KlasseBG"])
fig2 = px.violin(df, y=df["KlasseVABD"])
fig1.show()
fig2.show()
```

```python
fig = px.violin(df, x=df["BERVABD"], y=df["BEROEBG"], color= df["KlasseBG"], box=True, points="all",
            width=1600, height=1200)
fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()

```

```python
fig = px.violin(df, y= df["BEROEBG"], x=df["BERVABD"], color= df["KlasseVABD"],
                violinmode='overlay', # draw violins on top of each other
                # default violinmode is 'group' as in example above
                hover_data=df.columns, width=2000, height=1200)
fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
```
